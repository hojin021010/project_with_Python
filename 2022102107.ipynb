+ 리스트의 요소를 float type으로 변환하는 함수 생성 

```python
def typeListChange(input_list) :
    
    fixedlist =[]
    for i in input_list:
        fixedlist.append(float(i))
    return fixedlist
```

+ 리스트의 값에는 B, %등 str요소가 같이 있기 때문에 이를 분리하고 적절한 figure을 넣기 위한 함수 생성

```python
def typeListChangeStrToFloat(inputlist):
    
    i = 0 
    if type(inputlist[i]) == str:
        while i <= (len(inputlist)-1) :
            changebefore= inputlist[i]
            if changebefore[-1] == 'B':
                changeafter = float(changebefore[:-1])*1000000000
                inputlist[i] = changeafter 
            elif changebefore[-1] == '%':
                changeafter = float(changebefore[:-1])
                inputlist[i] = changeafter 
            else :
                return typeListChange(inputlist)
            i += 1
        return inputlist
    else :
        return typeListChange(inputlist)
```

#### 5. 분석 결과 도출

###### 상관계수 : 한 변수의 변화에 따른 다른 변수의 변화 정도와 방향을 예측하는 분석기법 
+ 양의 값이면 양의 상관관계, 음의 값이면 음의 상관관계라고 한다.
+ 양의 값이면 비례하는 관계거나 한 변수가 증가할 때 다른 변수 또한 같이 증가하는 관계이고,
+ 음의 값이면 반비례하는 관계거나 한 변수가 증가할 때 다른 변수는 감소하는 관계이다.

+ 상관관계 구하는 correlation 함수 정의
+ 리스트의 평균을 구하는 mean 함수 정의
+ 이들을 메소드로 가지는 'baseratecorr'이라는 class 정의
+ 리스트마다 value값을 가지는 기간이 다르기 때문에 상관관계를 구할 때 가장 최근의 자료를 이용한다. 이를 발췌하는 makesamelength 함수 정의
+ 리스트마다 value값을 가지는 기간이 다르기 때문에 이 기간을 맞추기 위해 리스트의 필요한 부분을 발췌해서 두 그래프를 겹쳐그려보며 상관관계를 눈으로 파악해보는 drawgraphbetween함수 정의

```python
def drawgraphbetween(list1, list2):
    
    if len(list1) == len(list2):
        x = list(range(1,len(list1)+1))
        graph1 = list1
        graph2 = list2
        plt.plot(x,graph1, 'r--', x , graph2, 'b--')
    else :
        if len(list1)> len(list2):
            x = list(range(1,len(list2)+1))
            cut = len(list2)
            graph1 = list1[-cut:-1]
            graph1.append(list1[-1])
            graph2 = list2
            plt.plot(x,graph1, 'r--', x , graph2, 'b--')
        else: 
            x = list(range(1,len(list1)+1))
            cut = len(list1)
            graph1 = list1
            graph2 = list2[-cut:-1]
            graph2.append(list2[-1])
            plt.plot(x,graph1, 'r--', x , graph2, 'b--')
            
        
class baseratecorr:
    
    def __init__(self, datalist1, datalist2):
        self.datalist1 = datalist1
        self.datalist2 = datalist2  
        
    def mean(self, data):
        result = 0
        len_data = len(data)
        for i in data:
            result += i
        return result / len_data 

    def makesamelength(self,list1,list2):
    
        if len(list1) == len(list2):
            pass
        else:
            if len(list1) > len(list2):
                list3 = list1[-len(list2):-1]
                list3.append(list1[-1])
                return list3
            else:
                list3 = list2[-len(list1):-1]
                list3.append(list2[-1])
                return list3
            
           
    def maincorrelation(self):
        if len(self.datalist1) == len(self.datalist2):
            return self.correlation(self.datalist1, self.datalist2)
        else:
            if len(self.datalist1) > len(self.datalist2):
                fixedlist = self.makesamelength(self.datalist1, self.datalist2)
                return self.correlation(fixedlist, self.datalist2)
            else:
                fixedlist = self.makesamelength(self.datalist1, self.datalist2)
                return self.correlation(self.datalist1, fixedlist)
        
        
    def correlation(self, data1, data2):
        meanData1 = self.mean(data1)
        meanData2 = self.mean(data2)
        up = 0
        for i in range(len(data1)):
            up += (data1[i]-meanData1) * (data2[i]-meanData2)
        bottom = 0
        data1Bottom = 0
        data2Bottom = 0
        for i in data1:
            data1Bottom += (i - meanData1)**2
        for i in data2:
            data2Bottom += (i - meanData2)**2
        bottom = math.sqrt(data1Bottom) * math.sqrt(data2Bottom)
        return up/bottom
```

1. 금리와 물가간의 상관관계 분석 : 기준금리와 소비자물가간의 상관관계

#### 6. 결론

1. 금리를 인상하면 물가가 오른다. 즉, 금리와 물가는 양의 상관관계를 가지고 있다.
    - 금리와 물가간의 상관관계를 구하기 이전, 생산자 물가지수와 소비자 물가지수간의 상관관계가 0.8311408259728383로 매우 크게 나타났다는 것을 알 수 있다. 
    - 앞선 금리와 물가간의 상관관계를 구했을 때, 0.8184734870529489라는 매우 높은 양의 상관관계지수를 보였다. 
    - 금리를 인상하면 물가가 오른다는 가설과 일치한다.

2. 금리를 인상하면 환율이 내려간다. 즉 금리와 환율은 음의 상관관계를 가지고 있다.
    - 예상과 다르게 금리와 환율의 상관과계지수가 0.9265763443555075라는 매우 높은 양의 상관관계로 나타났다.
    - 가설과 일치하지 않는 요인은 다음과 같다.
         - 먼저 금리와 환율간 어떻게 통상적으로 음의 상관관계를 가지고 있는지 알 필요가 있다.
         - 금리가 높아지면, 상대적으로 주변국가보다 은행 예금 이자율이 높아지기 때문에, 외환이 유입된다. 외환이 유입되면 환율이 하락하는 구조다.
         - 그렇지만, 우리가 금리를 인상하고 있지만, 미국의 기준금리도 꾸준히 증가하고있고, 우리나라의 기준금리보다 높기때문에 이러한 논리라면 외환이 빠져나가야한다. 따라서 환율이 계속 올라간다고 이해할 수도 있다.
    - 금리와 환율은 상관관계의 위치이지, 인과관계라고 볼 수는 없기 때문에, 이 두 수치 이외의 다른 요인에 의해 가설과 다른 값이 도출되었다.

 3. 금리를 인상하면 외환 보유액이 증가한다. 즉 금리와 외환 보유액은 양의 상관관계를 가지고 있다.
    - 예상과 다르게 -0.9773172809968912라는 매우 높은 음의 상관관계 수치가 나왔다. 이 역시 (2)의 금리와 환율간의 상관관계가 예상과 다르게 나온 것과 같게 해석할 수 있다. 
    - 즉, 금리와 외환 보유액또한 상관관계의 위치이지, 인과관계라고 볼 수는 없기 때문에, 이 두 수치 이외의 다른 요인에 의해 가설과 다른 값이 도출되었다. 

 4. 환율이 내려가면 수출이 감소하고 수입이 증가한다. / 환율이 올라가면 수출이 증가하고 수입이 감소한다. 즉, 환율은 수출과 양의 상관관계, 수입과 음의 상관관계를 가지고 있다.
    - 환율과 수출
        - 환율과 수출간의 상관계수가 -0.7619801579754716로 강한 음의 상관관계를 보인다고 할 수 있다.
        - 환율이 오르면 수출이 증가한다는 가설과 일치하지 않는데, 이는 두 수치 이외에 역시 다른 요인을 생각해보아야한다. 수출이 감소한 원인에는 2021년에는 코로나사태로 인해 전세계적인 수출 감소, 그리고 2022년에는 2022년 2월에 발발한 우크라이나 전쟁으로 인한 수출 감소를 꼽을 수 있다. 이렇게 두 수치만으로 비교했을 때는 예상과 다른 결과가 나왔지만, 다른 요인들을 살펴봐야한다.
        - 환율과 수출 또한 상관관계의 위치이지, 인과관계라고 볼 수는 없기 때문에, 이 두 수치 이외의 다른 요인에 의해 가설과 다른 값이 도출되었다.
    - 환율과 수입
        - 환율과 수입간의 상관계수가 -0.3307148074241634로 음의 상관관계를 보였다.즉, 환율이 올라가면 수입이 줄어든다는 처음 가설과 일치하는 것인데, 이 또한 결과론적으로는 맞지만, 단순히 환율의 문제로 볼 수는 없다. 수출과 마찬가지로 2021년에는 코로나, 2022년에는 우크라이나 전쟁으로 인해 수입, 수출이 줄어든 것이므로 이 데이터가 확실하다고 할 수는 없을 것이다.
    - 수출과 수입
        - 다른 예시를 살펴보아야한다는 것이 분명히 드러난 예시가 수출과 수입의 상관관계이다. 예상대로라면 음의 값을 보여야하는데, 0.6999479119743793의 양의 상관관계를 나타내고 있다. 이는 두 수치 모두 같이 줄어들거나, 같이 증가하고 있다는 뜻으로, 환율 이외의 요인에 의해 같이 감소하고있다는 것을 보여준다.

5. 금리를 인상하면 경기가 안정된다.
+ 금리 인상과 경기 안정간의 순서를 간략하게 나타내면 다음과 같다.
+ 금리가 높아진다 -> 외국 자본 유입 -> 환율 하락 -> 외국 자본 유출 -> 환율 다시 상승 -> 물가 상승 -> 수출 유리 -> 수출 증가 -> 환율 하락 -> 물가 안정

이를 증명하기 위해서는 앞서 가설을 정의한 4가지가 모두 만족해야한다. 그래야 금리인상으로 인한 물가 상승에도 불구하고 금리를 인상하는 것이 효과적이다.
1. 금리와 환율간의 음의 상관관계 
2. 금리와 외국 자본간의 양의 상관관계
3. 환율과 수출간의 양의 상관관계 
4. 환율과 수입간의 음의 상관관계

그렇지만 앞선 1,2,3,4중 4를 제외하고는 모두 가설과 다른 값이 도출되었다. 이는 금리인상만으로는 경기안정을 도모할 수 없다는 결론에 이른다. 경기 안정에는 금리 뿐 아니라, 세계 경제상황, 기후환경, 식량상태 등 여러가지 요인이 복합적으로 원인이 된다. 따라서 금리 인상과 경기 안정간에 상관관계가 있을지언정, 금리 인상이 경기 안정으로 바로 이어지는 인과관계는 될 수 없다는 결론에 다다른다.

#### 10. (3)을 (4)로 변환하는 등의 작업을 위하여 본인이 직접 개발한 Python 소스코드 원본

```python
import math
import pandas as pd 
import matplotlib.pylab as plt 

baserate = pd.read_csv("baserate.csv")
consumer_price_index = pd.read_csv("consumer_price_index.csv")
exchangerate = pd.read_csv("exchangerate.csv")
export = pd.read_csv("export.csv")
foreign_exchange_reserves = pd.read_csv("foreign_exchange_reserves.csv")
import_ = pd.read_csv("import_.csv")
industrial_production = pd.read_csv("industrial_production.csv")
producer_price_index = pd.read_csv("producer_price_index.csv")

def typeListChange(input_list) :
    
    fixedlist =[]
    for i in input_list:
        fixedlist.append(float(i))
    return fixedlist

def typeListChangeStrToFloat(inputlist):
    
    i = 0 
    if type(inputlist[i]) == str:
        while i <= (len(inputlist)-1) :
            changebefore= inputlist[i]
            if changebefore[-1] == 'B':
                changeafter = float(changebefore[:-1])*1000000000
                inputlist[i] = changeafter 
            elif changebefore[-1] == '%':
                changeafter = float(changebefore[:-1])
                inputlist[i] = changeafter 
            else :
                return typeListChange(inputlist)
            i += 1
        return inputlist
    else :
        return typeListChange(inputlist)
    
baserate1 = pd.read_csv("baserate.csv", names = ["date","figure","difference"])
baserateFigure = baserate1["figure"]
baserateFigure_Val = baserateFigure.values
baserateFigure_ValList =baserateFigure_Val.tolist() 
baserateFigure_ValList = baserateFigure_ValList[1:]
baserateList = baserateFigure_ValList

baserateList = typeListChangeStrToFloat(baserateList)

consumer_price_index2 = pd.read_csv("consumer_price_index.csv", names = ["date", "figure","difference"])
consumer_price_indexFigure = consumer_price_index2["figure"]
consumer_price_indexFigure_Val = consumer_price_indexFigure.values
consumer_price_indexFigure_ValList = consumer_price_indexFigure_Val.tolist()
consumer_price_indexFigure_ValList = consumer_price_indexFigure_ValList[1:]
cpiList = consumer_price_indexFigure_ValList

cpiList = typeListChangeStrToFloat(cpiList)

import_2 = pd.read_csv("import_.csv", names = ["date", "figure","difference"])
import_2Figure = import_2["figure"]
import_2Figure_Val = import_2Figure.values
import_2Figure_ValList = import_2Figure_Val.tolist()
import_2Figure_ValList = import_2Figure_ValList[1:]
importList = import_2Figure_ValList

importList = typeListChangeStrToFloat(importList)

export2 = pd.read_csv("export.csv", names = ["date", "figure","difference"])
export2Figure = export2["figure"]
export2Figure_Val = export2Figure.values
export2Figure_ValList = export2Figure_Val.tolist()
export2Figure_ValList = export2Figure_ValList[1:]
exportList = export2Figure_ValList

exportList = typeListChangeStrToFloat(exportList)

exchangerate2 = pd.read_csv("exchangerate.csv",names = ["date", "figure","difference"])
exchangerate2Figure = exchangerate2["figure"]
exchangerate2Figure_Val = exchangerate2Figure.values
exchangerate2Figure_ValList = exchangerate2Figure_Val.tolist()
exchangerate2Figure_ValList = exchangerate2Figure_ValList[1:]
exchangerateList = exchangerate2Figure_ValList

exchangerateList = typeListChangeStrToFloat(exchangerateList)

foreign_exchange_reserves2 = pd.read_csv("foreign_exchange_reserves.csv", names = ["date","figure","difference"])
foreign_exchange_reserves2Figure = foreign_exchange_reserves2["figure"]
foreign_exchange_reserves2Figure_Val = foreign_exchange_reserves2Figure.values
foreign_exchange_reserves2Figure_ValList = foreign_exchange_reserves2Figure_Val.tolist()
foreign_exchange_reserves2Figure_ValList = foreign_exchange_reserves2Figure_ValList[1:]
foreign_exchange_reservesList = foreign_exchange_reserves2Figure_ValList

foreign_exchange_reservesList = typeListChangeStrToFloat(foreign_exchange_reservesList)

producer_price_index2 = pd.read_csv("producer_price_index.csv", names = ["date","figure","difference"])
producer_price_index2Figure = producer_price_index2["figure"]
producer_price_index2Figure_Val = producer_price_index2Figure.values
producer_price_index2Figure_ValList = producer_price_index2Figure_Val.tolist()
producer_price_index2Figure_ValList = producer_price_index2Figure_ValList[1:]
producer_price_indexList = producer_price_index2Figure_ValList

producer_price_indexList = typeListChangeStrToFloat(producer_price_indexList)

industrial_production2 = pd.read_csv("industrial_production.csv", names = ["date","figure","difference"])
industrial_production2Figure = industrial_production2["figure"]
industrial_production2Figure_Val = industrial_production2Figure.values
industrial_production2Figure_ValList = industrial_production2Figure_Val.tolist()
industrial_production2Figure_ValList = industrial_production2Figure_ValList[1:]
industrial_productionList = industrial_production2Figure_ValList

industrial_productionList = typeListChangeStrToFloat(industrial_productionList)

def drawgraphbetween(list1, list2):
    
    if len(list1) == len(list2):
        x = list(range(1,len(list1)+1))
        graph1 = list1
        graph2 = list2
        plt.plot(x,graph1, 'r--', x , graph2, 'b--')
    else :
        if len(list1)> len(list2):
            x = list(range(1,len(list2)+1))
            cut = len(list2)
            graph1 = list1[-cut:-1]
            graph1.append(list1[-1])
            graph2 = list2
            plt.plot(x,graph1, 'r--', x , graph2, 'b--')
        else: 
            x = list(range(1,len(list1)+1))
            cut = len(list1)
            graph1 = list1
            graph2 = list2[-cut:-1]
            graph2.append(list2[-1])
            plt.plot(x,graph1, 'r--', x , graph2, 'b--')
            
        
class baseratecorr:
    
    def __init__(self, datalist1, datalist2):
        self.datalist1 = datalist1
        self.datalist2 = datalist2  
        
    def mean(self, data):
        result = 0
        len_data = len(data)
        for i in data:
            result += i
        return result / len_data 

    def makesamelength(self,list1,list2):
    
        if len(list1) == len(list2):
            pass
        else:
            if len(list1) > len(list2):
                list3 = list1[-len(list2):-1]
                list3.append(list1[-1])
                return list3
            else:
                list3 = list2[-len(list1):-1]
                list3.append(list2[-1])
                return list3
            
           
    def maincorrelation(self):
        if len(self.datalist1) == len(self.datalist2):
            return self.correlation(self.datalist1, self.datalist2)
        else:
            if len(self.datalist1) > len(self.datalist2):
                fixedlist = self.makesamelength(self.datalist1, self.datalist2)
                return self.correlation(fixedlist, self.datalist2)
            else:
                fixedlist = self.makesamelength(self.datalist1, self.datalist2)
                return self.correlation(self.datalist1, fixedlist)
        
        
    def correlation(self, data1, data2):
        meanData1 = self.mean(data1)
        meanData2 = self.mean(data2)
        up = 0
        for i in range(len(data1)):
            up += (data1[i]-meanData1) * (data2[i]-meanData2)
        bottom = 0
        data1Bottom = 0
        data2Bottom = 0
        for i in data1:
            data1Bottom += (i - meanData1)**2
        for i in data2:
            data2Bottom += (i - meanData2)**2
        bottom = math.sqrt(data1Bottom) * math.sqrt(data2Bottom)
        return up/bottom
```
